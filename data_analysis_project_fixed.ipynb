# 📊 BigMart Sales Dataset Analysis (EDA Project)

```python
# Step 1: Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

```python
# Step 2: Load dataset

df = pd.read_csv("Test.csv")  # change file name if needed

print("Columns in dataset:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)
```

```python
# Step 3: Dataset Information
print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())
```

```python
# Step 4: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 5.1: Handle Missing Values
# Fill Item_Weight with mean
if 'Item_Weight' in df.columns:
    df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)

# Fill Outlet_Size with mode
if 'Outlet_Size' in df.columns and not df['Outlet_Size'].mode().empty:
    df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Step 5.2: Fix Item_Fat_Content Values
if 'Item_Fat_Content' in df.columns:
    df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
        'reg': 'Regular',
        'low fat': 'Low Fat',
        'LF': 'Low Fat',
        'Lowfat': 'Low Fat',
        'lf': 'Low Fat',
        'Reg': 'Regular'
    })
    print(df['Item_Fat_Content'].value_counts())
```

```python
# Step 6: Distribution of Item MRP
plt.figure(figsize=(8,5))
sns.histplot(df['Item_MRP'], bins=20, kde=True)
plt.title("Distribution of Item MRP")
plt.show()
```

```python
# Step 7: Most Common Item Type
plt.figure(figsize=(10,5))
sns.countplot(y='Item_Type', data=df, order=df['Item_Type'].value_counts().index)
plt.title("Count of Item Types")
plt.show()
```

```python
# Step 8: Fat Content Distribution
sns.countplot(x='Item_Fat_Content', data=df)
plt.title("Item Fat Content Distribution")
plt.show()
```

```python
# Step 9: Outlet Type Distribution
sns.countplot(x='Outlet_Type', data=df)
plt.title("Outlet Type Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 10: Outlet Location Type Distribution
sns.countplot(x='Outlet_Location_Type', data=df)
plt.title("Outlet Location Type Distribution")
plt.show()
```

```python
# Step 11: Item MRP by Outlet Type
plt.figure(figsize=(8,5))
sns.boxplot(x='Outlet_Type', y='Item_MRP', data=df)
plt.xticks(rotation=45)
plt.title("Item MRP by Outlet Type")
plt.show()
```

```python
# Step 12: Item Visibility Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Item_Visibility'], bins=20, kde=True)
plt.title("Item Visibility Distribution")
plt.show()
```

```python
# Step 13: Relationship Between Item MRP and Visibility
plt.figure(figsize=(8,5))
sns.scatterplot(x='Item_Visibility', y='Item_MRP', data=df)
plt.title("Item Visibility vs Item MRP")
plt.show()
```

```python
# Step 14: Average MRP by Item Type
avg_mrp = df.groupby('Item_Type')['Item_MRP'].mean().sort_values()

plt.figure(figsize=(10,6))
avg_mrp.plot(kind='barh')
plt.title("Average MRP by Item Type")
plt.show()
```

```python
# Step 15: Outlet Establishment Year Analysis
plt.figure(figsize=(8,5))
sns.countplot(x='Outlet_Establishment_Year', data=df)
plt.xticks(rotation=45)
plt.title("Outlet Establishment Year Distribution")
plt.show()
```

```python
# Step 16: Heatmap (Correlation Matrix)
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
